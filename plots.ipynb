```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv('cancer.csv')


print(df.info())

df = df.drop(['index', 'Patient Id'], axis=1)

label_encoder = LabelEncoder()
df['Level'] = label_encoder.fit_transform(df['Level'])
```

```python
print(df.isnull().sum())
```

```python
df.hist(bins=10, figsize=(20, 15))
plt.suptitle('Histograms of Features')
plt.show()
```

```python
df.plot(kind='box', subplots=True, layout=(5, 5), figsize=(20, 15))
plt.suptitle('Box Plots of Features')
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
df['Level'].value_counts().plot(kind='bar')
plt.title('Distribution of Levels')
plt.xlabel('Level')
plt.ylabel('Count')
plt.show()
```

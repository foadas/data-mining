```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np
from scipy import stats

df = pd.read_csv('cancer.csv')

data = df.drop(['index', 'Patient Id','Level'], axis=1)


```

```python
z_scores = stats.zscore(data)
z_scores_df = pd.DataFrame(z_scores, columns=data.columns)
threshold = 3
outliers = (z_scores_df > threshold) | (z_scores_df < -threshold)
outlier_rows = df[outliers.any(axis=1)]


print("Outlier Data:")
print(outlier_rows)

output_path = 'outlier_data_z_score.csv' 
outlier_rows.to_csv(output_path, index=False)
```

```python
numeric_data = df.select_dtypes(include=[float, int])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)


kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(scaled_data)
df['Cluster'] = clusters


distances = kmeans.transform(scaled_data)
min_distances = np.min(distances, axis=1)


threshold = np.percentile(min_distances, 95)
outliers = df[min_distances > threshold]


print("Outlier Data:")
print(outliers)


output_path = 'outlier_data_by_clustering.csv'
outliers.to_csv(output_path, index=False)


df_cleaned = df[min_distances <= threshold]


df_cleaned = df_cleaned.drop(columns=['Cluster'])


cleaned_output_path = 'cleaned_cancer.csv'
df_cleaned.to_csv(cleaned_output_path, index=False)

```

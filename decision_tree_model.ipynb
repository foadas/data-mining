```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from model_evaluation_methods import *
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

df = pd.read_csv('cleaned_cancer.csv')
df = df.drop(['index', 'Patient Id'], axis=1)


X = df.drop('Level', axis=1)
y = df['Level']
```

```python
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)
```

```python
selector = SelectFromModel(DecisionTreeClassifier(random_state=42))
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

decision_tree_model_fs = DecisionTreeClassifier(random_state=42)
decision_tree_model_fs.fit(X_train_selected, y_train)

y_pred_fs = decision_tree_model_fs.predict(X_test_selected)

conf_matrix_fs = confusion_matrix(y_test, y_pred_fs)
error_rate_fs = calculate_error_rate(conf_matrix_fs)
sensitivity_fs,precision_fs, f_measure_fs = calculate_metrics(y_test, y_pred_fs)
accuracy_fs = accuracy_score(y_test, y_pred_fs)


print("decision tree model Evaluation for Feature Selection:")
print("Accuracy:", accuracy_fs)
print("Error Rate:", error_rate_fs)
print("Sensitivity (Recall):", sensitivity_fs)
print("Precision:", precision_fs)
print("F-measure:", f_measure_fs)

plot_roc_curve_multiclass(y_test, decision_tree_model_fs.predict_proba(X_test_selected), decision_tree_model_fs.classes_,'decision_tree_model_fs')

y_train_pred_fs = decision_tree_model_fs.predict(X_train_selected)

train_accuracy_fs = accuracy_score(y_train, y_train_pred_fs)

print("Training vs Testing Accuracy Comparison:")

print("Feature Selection - Training Accuracy:", train_accuracy_fs, "Testing Accuracy:", accuracy_fs)

cv_scores_fs = cross_val_score(decision_tree_model_fs, X_train_selected, y_train, cv=5, scoring='accuracy')

print("Cross-Validation Scores:")

print("Feature Selection - Mean CV Accuracy:", np.mean(cv_scores_fs))

plot_learning_curve(decision_tree_model_fs, "Learning Curve - Decision tree (Feature Selection)", X_train_selected, y_train)

plt.show()
```

```python
pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

decision_tree_model_pca = DecisionTreeClassifier(random_state=42)
decision_tree_model_pca.fit(X_train_pca, y_train)

y_pred_pca = decision_tree_model_pca.predict(X_test_pca)

conf_matrix_pca = confusion_matrix(y_test, y_pred_pca)
error_rate_pca = calculate_error_rate(conf_matrix_pca)
sensitivity_pca, precision_pca, f_measure_pca = calculate_metrics(y_test, y_pred_pca)
accuracy_pca = accuracy_score(y_test, y_pred_pca)

print("\ndecision tree model Evaluation for PCA:")
print("Accuracy:", accuracy_pca)
print("Error Rate:", error_rate_pca)
print("Sensitivity (Recall):", sensitivity_pca)
print("Precision:", precision_pca)
print("F-measure:", f_measure_pca)

plot_roc_curve_multiclass(y_test, decision_tree_model_pca.predict_proba(X_test_pca), decision_tree_model_pca.classes_,'decision_tree_model_pca')

y_train_pred_pca = decision_tree_model_pca.predict(X_train_pca)

train_accuracy_pca = accuracy_score(y_train, y_train_pred_pca)

print("Training vs Testing Accuracy Comparison:")

print("PCA - Training Accuracy:", train_accuracy_pca, "Testing Accuracy:", accuracy_pca)

cv_scores_pca = cross_val_score(decision_tree_model_pca, X_train_pca, y_train, cv=5, scoring='accuracy')

print("Cross-Validation Scores:")

print("PCA - Mean CV Accuracy:", np.mean(cv_scores_pca))

plot_learning_curve(decision_tree_model_pca, "Learning Curve - Decision tree (PCA)", X_train_pca, y_train)

plt.show()
```
